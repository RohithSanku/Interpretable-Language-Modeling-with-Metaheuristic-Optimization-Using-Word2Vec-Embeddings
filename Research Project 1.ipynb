```python
import string
import re
import nltk.corpus
from nltk.corpus import ptb
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
nltk.download('punkt')
```

```python
from nltk.corpus import treebank
nltk.download('treebank')
```

```python

# Example sentences (replace this with your actual sentences)
sentences = treebank.sents()

tokenized_sentences = [word_tokenize(" ".join(sentence)) for sentence in sentences]
print(tokenized_sentences)
```

```python
def extract_alphanumeric(text):
    # Find all alphanumeric characters and underscores
    return re.findall(r'[a-zA-Z]+', text)


cleaned_text = [extract_alphanumeric(" ".join(sentence)) for sentence in tokenized_sentences]

print(cleaned_text)
```

```python
input_sequences = []
target_words = []

for sentence in tokenized_sentences:
    for i in range(3, len(sentence)):
        # X is the sequence of the first 3 words
        X = sentence[i-3:i]
        # Y is the next word
        Y = sentence[i]

        input_sequences.append(X)
        target_words.append(Y)
print("Sample Input Sequence:", input_sequences[0])
print("Sample Target Word:", target_words[0])
```

```python
import numpy as np
# Step 3: Train Word2Vec model
model = Word2Vec(sentences=tokenized_sentences, vector_size=300, window=5, min_count=1, workers=4)

# Step 4: Convert words to vectors using Word2Vec model
X_vectors = [model.wv[word] for word in input_sequences]
Y_vectors = [model.wv[word] for word in target_words if word in model.wv]


# Convert lists to numpy arrays
X_array = np.array(X_vectors)
Y_array = np.array(Y_vectors)
print(Y_array.shape)
```

```python
# Initialize A, M, and C randomly
A = np.random.rand(1, 3)
M = np.random.rand(3, 3)
C = np.random.rand(3, 300)
```

```python
#calculate y'
y_p=np.zeros((72206,300))
cosine=np.zeros((72206))
from numpy.linalg import norm
for i in range(72206):
    y_p[i,:]=np.matmul(A,np.matmul(M,X_array[i,:,:])+C)
#calculate error
    #print(y_p.shape)
    cosine[i] = np.dot(y_p[i,:],Y_array[i,:])/(norm(y_p[i,:])*norm(Y_array[i,:]))
print(sum(cosine))
```

```python
best_fit = 50
fit=np.zeros((72206))
p=10
for iteration in range(100000):  # Number of iterations can be set as needed
  for i in range(72206):
    y_p[i,:]=pow(np.matmul(A,np.matmul(M,X_array[i,:,:])+C),p)
    fit[i] = np.dot(y_p[i,:],Y_array[i,:])/(norm(y_p[i,:])*norm(Y_array[i,:]))
  print(sum(fit),best_fit)
  if sum(fit) > best_fit:
    best_fit = sum(fit)
    A_best=A
    M_best=M
    C_best=C
    p_best=p
    t=[f for f in fit if f ==1]
    print(best_fit,iteration,sum(t))
  A = (np.random.rand(1, 3)+A_best)/2
  M = (np.random.rand(3, 3)+M_best)/2
  C = (np.random.rand(3, 300)+C_best)/2
  p=int((np.random.rand()*5+p_best)/2)
```
